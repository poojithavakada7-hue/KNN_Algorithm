```python
# Import Libraries
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Training Data
X_train = np.array([[1], [2], [3], [6], [7], [8]])
y_train = np.array([0, 0, 0, 1, 1, 1])

# Create Model with K = 3
model = KNeighborsClassifier(n_neighbors=3)

# Train Model
model.fit(X_train, y_train)

# Test Data (New Student studied 4 hours)
X_test = np.array([[7]])

# Predict
prediction = model.predict(X_test)

print("Predicted Class (0=Fail, 1=Pass):", prediction[0])
```

```python
# Testing on multiple values
X_test2 = np.array([[2], [5], [7]])
y_test2 = np.array([0, 1, 1])

predictions = model.predict(X_test2)

print("Predictions:", predictions)
print("Accuracy:", accuracy_score(y_test2, predictions))
```
